```python
#Imports
import pandas as pd
import numpy as np
import seaborn as sb
import sqlite3
import sqlalchemy
import matplotlib.pyplot as plt
from datetime import datetime as dt
```

```python
#data set
conn = sqlite3.connect("database.sqlite")
```

```python
#User table inport and cleaning, renamed to 'age' to prevent confusion
users = pd.read_sql_query("select * from user;", conn)
age = users.drop(columns=["first_name", "last_name", "city", 
                          "competitions", "occupation", "sponsor1",
                         "sponsor2", "sponsor3", "best_area",
                         "worst_area", "guide_area", "interests",
                          "presentation", "deactivated", "anonymous"])
```

```python
#Importing grade and cleaning
grade = pd.read_sql_query("select * from grade;", conn)
grade_reduced = grade.drop(columns=["score", "fra_routes_input", "fra_routes_selector", 
                          "fra_boulders", "fra_boulders_input", "fra_boulders_selector",
                         "usa_routes_input", "usa_routes_selector", "usa_boulders_input",
                         "usa_boulders_selector"])
```

```python
#Importing ascent table slightly different becasue of size
curs = conn.cursor()
curs.execute("select id, user_id, grade_id, year from ascent; ")
data = curs.fetchall()
ascent = pd.DataFrame(data, columns= ['id', 'user_id', 'grade_id', 'year'])
```

```python
#Changing column "id" to "grade_id" for merging 
grade1 = grade_reduced.rename(columns={"id": "grade_id"})
```

```python
#renaming "id" to "user_id" for merging
age1 = age.rename(columns={"id": "user_id"})
```

```python
# Merging users(age) and ascents 
climbers = pd.merge(age1, ascent, on='user_id', how='inner')
```

```python
# Merging all tables I need together (climbers & grade1)
climber = pd.merge(climbers, grade1, on='grade_id', how='inner')
```

```python
#Decided to drop all nan values becasue I feel 
#it shouldn't comprimise data(only dropped about 1/4 of data)
climb = climber.replace('None', np.NaN).dropna()
```

```python
#pushing the 'birth' index to just be year
climb['birth'] = pd.to_datetime(climb['birth'], errors='coerce').dt.year
climb = climb.dropna()
```

```python
#removing all nonsense birthdays; i chose 1918 because there was one questionable individual claiming to be born in 1918
climb['birth'] = climb['birth'].astype(int)
climb = climb[climb['birth'] >= 1918]
```

```python
#had to make an age when climbed route to get an accurate representation of ability
age_of_climb = climb['year'] - climb['birth']
```

```python
#Added age_of_climb onto table
climb['age_of_climb'] = age_of_climb
```

```python
#dropping all non relevant ages ie. someone said they were born in the year 1000
climb = climb.drop (climb[climb['age_of_climb'] <=0].index)
climb['age_of_climb'].nunique()
```

```python
#Subsetting "grade_id" for skill level
climb_low = climb[(climb['grade_id'] <= 34)]
climb_mid = climb[(climb['grade_id'] <= 52) & (climb['grade_id'] >= 35)]
climb_high = climb[(climb['grade_id'] <= 66) & (climb['grade_id'] >= 53)]
climb_advanced = climb[(climb['grade_id'] > 67)]
```

```python
#putting all the above graphs toghther for compairson.
fig, axs = plt.subplots(2, 2)
fig.set_size_inches(15, 8)

axs[0, 0].hist((climb_low['age_of_climb']), bins=50, color='#67BC44')
axs[0, 0].set_title('Climbs Completed in Beginner Category')
axs[0, 1].hist((climb_mid['age_of_climb']), bins=50, color='#E5Eb0E')
axs[0, 1].set_title('Climbs Completed in Intermediate Category')
axs[1, 0].hist((climb_high['age_of_climb']), bins=50, color='#EB1D23') 
axs[1, 0].set_title('Climbs Completed in Advanced Category')
axs[1, 1].hist((climb_advanced['age_of_climb']), bins=50, color='#AB4B9B')
axs[1, 1].set_title('Climbs Completed in Pro Category')

for ax in axs.flat:
    ax.set(xlabel='Age of Climber', ylabel='Number of Climbs')
    
    
    
#Highlight differences-----------------------------------------------------

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
plt.savefig('all_grades.png', dpi=400)
```

```python
#trying out a different method of above subsetting for efficiency
def to_bracket (val):
    if (val <= 34):
        return 'Beginner Skill'
    elif (val <= 52) and (val >= 35):
        return 'Intermediate Skill'
    elif (val <= 66) and (val >= 53):
        return 'Advanced Skill'
    elif (val > 67):
        return 'Pro Skill'
test_climb = climb['grade_id'].apply(to_bracket)
```

```python
# some times I need to run this graph twice to get it to display correctly
colors = ['#67BC44','#E5Eb0E', '#EB1D23','#AB4B9B']
plt.style.use('fivethirtyeight')

for i,grade in enumerate (['Beginner Skill', 'Intermediate Skill', 'Advanced Skill', 'Pro Skill']):
    climb[test_climb == grade]['age_of_climb'].value_counts().sort_index().plot.area(color=colors[i], label=grade, alpha=.5)


plt.rcParams['figure.figsize']=(25,15)
plt.title('Climber by Age', size=60)
plt.ylabel('Number of Climbs', size=60)
plt.xlabel('Climbers Age', size=60)
plt.yticks([0,125000])
plt.xticks(size=30)
# plt.xticks(range(0,99,10))
# plt.xticks([1,25,50,100])

plt.legend(prop={'size': 50})
plt.show()
```

```python
#Normalized graph
plt.style.use('fivethirtyeight')
colors = ['#67BC44','#E5Eb0E', '#EB1D23','#AB4B9B']

for i,grade in enumerate (['Beginner Skill', 'Intermediate Skill', 'Advanced Skill', 'Pro Skill']):
#     print (climb[climb['grade_id'] == grade]['age_of_climb'].value_counts(normalize=True))
    climb[test_climb == grade]['age_of_climb'].value_counts(normalize=True).sort_index().plot.area(color=colors[i], label=grade, alpha=.5)

plt.rcParams['figure.figsize']=(25,15)
plt.title('Climbers Normalized', size=60)
plt.ylabel('Number of Climbs', size=60)
plt.xlabel('Climbers Age', size=60)
plt.yticks([])
plt.xticks(size=30)
# plt.xticks(range(0,99,10))
# plt.xticks([1,25,50,100])

plt.legend(prop={'size': 50})
plt.show()
```
